## Modeling

Let's predict the daily closing price.

In practice, when the closing price comes in for each day, we can then add that datum to our model to aid us in predicting tomorrow's closing price.

I am going to reserve a week of closing prices for validation, and the week after that for testing. Note that these are 5-day weeks (business calendar.)

```python
train = expr_daily[:-10]['close']
validation = expr_daily[-10:-5]['close']
test = expr_daily[-5:]['close']
```

Before we get started, let's briefly visualize the time series.

```python
train.plot();
```

Here is a zoomed-in view of the last 10 weeks of the training set + the validation and test sets.

```python
train[-50:].plot()
validation.plot(c='r')
test.plot(c='g');
```

Here is a plot of the training data and its rolling average (with a window of 5 days.)

```python
train.plot()
train.rolling(window=5).mean().plot();
```

And here is the difference between the training data and its rolling average.

(This makes me think that the data may be a mixture model, maybe a random walk with drift.)

```python
(train - train.rolling(window=5).mean()).plot();
```

There is a clear overall downward trend. If this is a purely stochastic trend (based on the errors of the model at previous time steps), then we can remove the trend through differencing. If the trend turns out to be deterministic, then we need to regress a polynomial model on the time series with error removed.

This is what the first difference of the training data looks like:

```python
train.diff().plot();
```

So we will use a first order difference in our ARIMA model.

First, let's try a SARIMA model. To select the parameters for the model, we'll examine the autocorrelation and partial autocorrelation plots.

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
import matplotlib.pyplot as plt

fig, ax = plt.subplots(ncols=2, figsize=(12, 4))

fig.suptitle('Original Time Series')

ax[0].set_title('Auto-correlation')
plot_acf(train, ax=ax[0])
ax[1].set_title('Partial auto-correlation')
plot_pacf(train, ax=ax[1]);
```

The autocorrelation decreases smoothly while the partial autocorrelation drops off sharply after 2 lags.

ACF: the time series has a nearly perfect correlation with its lags up to 35 lags (and probably beyond.)

PACF: the correlation between the time series and its lags, corrected for mutual correlation with the lags inbetween, drops off sharply after the second lag. So, the time series is highly correlated with its first lag, and it is highly correlated with its second lag even after removing the correlation between the series and its first lag. But the correlation we observe between the time series and its later lags (the third lag onward) results mostly from correlation between the series and "in between" lags.

Because the error terms of the lags do not provide much information about the current value beyond two lags, we can try building a moving average model with q = 2.
